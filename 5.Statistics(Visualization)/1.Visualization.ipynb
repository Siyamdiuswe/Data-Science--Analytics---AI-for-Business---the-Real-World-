```python
import pandas as pd
import numpy as np

file_name = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv"
df = pd.read_csv(file_name)
df.head()
```

```python
print(df["alcohol"].min())
print(df["alcohol"].mean())
print(df["alcohol"].max())
```

```python
df.hist(column='alcohol', bins=15)
```

```python
# Let's get a bit more advanced and learn to customize your plots

df.hist(column='alcohol', bins=20, grid=False, figsize=(10,6), color='#96bb7c')
```

```python
import seaborn as sns

sns.distplot(df['alcohol'], bins=25)
```

```python
import seaborn as sns

sns.distplot(df['alcohol'], bins=25, kde=False, color='#534e52')
```

```python
df['alcohol'].value_counts().head()
```

```python
# Let's start with labeling our axis

import matplotlib.pyplot as plt

sns.displot(df.alcohol)
plt.xlabel('Alcohol Percentage')
plt.ylabel('Count')
plt.title("Alcohol Content")
plt.show()
```

```python
#Use sns.set_style() to control styling

sns.set_style("dark")
sns.histplot(df.alcohol, bins=15)
```
